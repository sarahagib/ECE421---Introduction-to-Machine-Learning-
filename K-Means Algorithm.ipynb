```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
cancer_data = dataset['data']
```

```python
def k_means(k,data):
    centroids = np.zeros((k,30))

    for i in range(k):
        temp = np.random.choice((len(data)))
        centroids[i] = data[temp]

    distance = np.zeros((k))
    assignments = np.zeros(len(data))

    x=0
    while x<100:
        for i in range(len(data)):
            for j in range(k):
                distance[j] = np.linalg.norm((centroids[j]-data[i]))
            assignments[i] = np.argmin(distance)

        for i in range(k):
            cluster = []
            for j in range(len(data)):
                if assignments[j] == i:
                    cluster.append(data[j])
            centroids[i] = np.mean(cluster,axis=0)
        x+=1
    return centroids,assignments
```

```python
def k_distortion(centroid, assignment, data):
  dist = 0
  sum = 0
  
  #calculate the difference and square it
  for i in range(len(centroid)):
    for j in range(len(data)):
      if assignment[j] == i: 
        diff = np.subtract(data[j],centroid[i])
        sqr = np.square(np.linalg.norm(diff)) 
        #squaring a number only
        sum +=sqr
  dist = 1/(len(data)) * sum
  return dist
```
